# Plotting

[aplpy](https://aplpy.github.io/) is a Python library for producing plots of astronomical data from FITS files. In this notebook we'll use the package to plot a simple single layer image, overlay contours and markers, and finally combine three FITS files at different wavelengths in to an RGB image.

First we load our typical libraries, as well as the `fits` sublibrary from Astropy and the `aplpy` library.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from astropy.io import fits
import aplpy
```

### Simple plotting

To create a canvas from an image use the `FITSFigure` function, passing a FITS file with an image in the primary HDU as the argument. To plot this canvas call the `show_grayscale` function to produce a grayscale plot. You can safely ignore the warnings given here.

```python
gc = aplpy.FITSFigure('../data/502nmos.fits')
gc.show_grayscale()
```

To plot this same image using a colorscale, use the `show_colorscale` function. We can also adjust other properties of the image from the canvas object. For example, to add a grid call the `add_grid` function. In the image below we also adjust the tick label font size.

```python
gc = aplpy.FITSFigure('../data/502nmos.fits')
gc.show_colorscale()
gc.add_grid()
gc.tick_labels.set_font(size='small')
```

### Adding Layers

aplpy allows you to overlay images with other data through markers or contours. To demonstrate, we will plot an image of M51 in the optical, overlay a radio map, and plot the locations of supernovae or supernovae remnants. 

Below we import `Pandas` and load a text file containing data on supernovae and supernovae remnants in M51. We then extract the RA and DEC information in to numpy arrays (aplpy does not currently accept Pandas data frames).

```python
import pandas as pd

objects = pd.read_csv('../data/objsearch.txt',sep='|',skiprows=24)

ra, dec = np.asarray(objects['RA(deg)']), np.asarray(objects['DEC(deg)'])
```

We can now plot the image. First load the optical data in to a canvas object, then plot with `show_colorscale`, specifying that we want a `gist_heat` colour map. We can then add a contour map of the radio data, specifying the colour and number of levels we wish to render. Finally, we can overlay the SN data we loaded above with `show_markers`, specifying the marker icon, edge and face colours, size (s) and transparency (alpha).

```python
m51 = aplpy.FITSFigure('../data/MESSIER_051-I-103aE-dss1.fits')
m51.show_colorscale(cmap='gist_heat')
m51.show_contour('../data/MESSIER_051-I-20cm-pkwb1984.fits', colors="white", levels=3)
m51.show_markers(ra, dec, edgecolor='green', facecolor='none', marker='*', s=100, alpha=0.8)
```

### Creating an RGB image

`aplpy` allows you to combine FITS datasets into an RGB image, which you can save out to a PNG or JPEG file. We'll demonstrate this with a three colour image of M17 from [spacetelescope.org](https://www.spacetelescope.org/projects/fits_liberator/m17data/). The three bands are OIII, H-$\alpha$ and SII, which correspond to blue, green and red respectively.

We first need to create a data cube containing our three FITS images. We do this with `make_rgb_cube`, passing a list of the filenames we wish to plot in the order red-green-blue, and a filename for the FITS cube we want to create. `make_rgb_cube` takes care of reprojecting the images in to the same plane using the coordinate data in each FITS file header.

Whilst we only specify a single file destination, in fact two FITS files are created: one with the full 3D data, and another with the suffix "_2d" added to the output name. This second file will be used later to plot the image.

```python
aplpy.make_rgb_cube(['../data/673nmos.fits','../data/656nmos.fits','../data/502nmos.fits'], '../data/nmod_cube.fits')
```

Using this data cube we can call `make_rgb_image` to create a PNG.

```python
aplpy.make_rgb_image('../data/nmod_cube.fits','../data/output/nmod.png')
```

To display the image we use the 2D FITS file create with `make_rgb_cube` to initialise a FITS figure. We then call `show_rgb` over this to plot the composite RGB image over the top.

```python
f = aplpy.FITSFigure('../data/nmod_cube_2d.fits')
f.show_rgb('../data/output/nmod.png')
```

By loading the 2D FITS file we provide information on the coordinate space of the image from the FITS header, information that is lost in the PNG conversion. 

Below we create the same image again, but add an inverse hyperbolic sine stretch to each colour layer in the call to `make_rgb_image`, which tends to flatten the image. After loading and plotting the PNG, we then add, position and label a *scalebar* to show the relative size of the nebula, and change the tick label format for clarity.

```python
aplpy.make_rgb_image('../data/nmod_cube.fits','../data/output/nmod.png', stretch_r='arcsinh', stretch_g='arcsinh', stretch_b='arcsinh')

f = aplpy.FITSFigure('../data/nmod_cube_2d.fits')
f.show_rgb('../data/output/nmod.png')

f.add_scalebar(5/60.)
f.scalebar.set_corner('top')
f.scalebar.set_length(17/600.)
f.scalebar.set_label('0.1 parsec')

f.tick_labels.set_xformat('hhmmss')
f.tick_labels.set_yformat('hhmmss')
```
